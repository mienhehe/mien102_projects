# Check data imbalance

```python
loan_acceptance_count = pd.DataFrame(dt['Personal Loan'].value_counts()).reset_index()
loan_acceptance_count.columns = ['Labels', 'Personal Loan']
loan_acceptance_count
```

```python
pie_labels = loan_acceptance_count['Labels']
pie_labels = ['Not Accepted' if x == 0 else 'Accepted' for x in pie_labels]
pie_data = loan_acceptance_count['Personal Loan'] 
explode = (0, 0.15) 
wp = { 'linewidth' : 1, 'edgecolor' : '#000000' }

def func(pct, allvalues): 
    absolute = int(np.round(pct / 100.*np.sum(allvalues)))
    return "{:.1f}%\n({:d})".format(pct, absolute)

fig, ax = plt.subplots(figsize =(10, 6))

ax.pie(pie_data,  
       autopct = lambda pct: func(pct, pie_data), 
       explode = explode,  
       labels = pie_labels, 
       shadow = True, 
       startangle = 70, 
       wedgeprops = wp)

ax.axis('equal') 
plt.title('Personal Loan Acceptance Percentage', size=22)
plt.show();
```

# Comment:
The dataset is unbalanced.

# Direction
There are algorithms that give good results on severely imbalanced datasets but are less efficient on unbalanced datasets.
imbalance and vice versa. Therefore, we should not trust one algorithm but must extend and test the model on many different algorithms. Let's test train a series of algorithms on the train set and check the accuracy of the prediction on the test set.
When severe data imbalance occurs, the use of accuracy as a measure of model evaluation
are often inefficient because most of them have very high accuracy because a random model predicts mostly labels
belonging to the majority group will bring results close to 100% leading to confusion. Then we can consider some alternative metrics such as precision, recall, f1-score (they are more focused on assessing accuracy on the minority group, which we want to forecast more accurately than majority group.)

# Split

```python
from sklearn.model_selection import train_test_split

X = dt.drop(columns='Personal Loan')

y= dt["Personal Loan"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)


X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train2 = pd.DataFrame(sc_X.fit_transform(X_train))
X_train2.columns = X_train.columns.values
X_train2.index = X_train.index.values
X_train = X_train2
X_test2 = pd.DataFrame(sc_X.transform(X_test))
X_test2.columns = X_test.columns.values
X_test2.index = X_test.index.values
X_test = X_test2
```

# Run classification models

# Random Forest model

```python
from sklearn.ensemble import RandomForestClassifier
model1= RandomForestClassifier(n_estimators=500,random_state=0)
model1.fit(X_train,y_train)

```

```python
from sklearn.metrics import confusion_matrix,classification_report
import sklearn.metrics as metrics
from sklearn.metrics import accuracy_score
y_pred1 = model1.predict(X_test)
print(accuracy_score(y_test,y_pred1))
```

```python

print(classification_report(y_test,y_pred1 ))
metrics.plot_confusion_matrix(model1, X_test, y_test)
```

```python
from sklearn.metrics import auc, roc_auc_score, roc_curve
model1.fit(X_train, y_train) 
probs = model1.predict_proba(X_test) 
probs = probs[:, 1] 
classifier_roc_auc = accuracy_score(y_test, y_pred1 )
rf_fpr, rf_tpr, rf_thresholds = roc_curve(y_test, model1.predict_proba(X_test)[:,1])
plt.figure(figsize=(8, 4))
plt.plot(rf_fpr, rf_tpr, 
label='RandomForest (area = %0.2f)' % classifier_roc_auc)
plt.plot([0,1], [0,1],label='Random Classifier ' 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.ylabel('True Positive Rate \n',horizontalalignment="center",
fontstyle = "normal", fontsize = "medium", 
fontfamily = "sans-serif")
plt.xlabel('\nFalse Positive Rate \n',horizontalalignment="center",
fontstyle = "normal", fontsize = "medium", 
fontfamily = "sans-serif")
plt.title('ROC Graph \n',horizontalalignment="center", 
fontstyle = "normal", fontsize = "22", 
fontfamily = "sans-serif")
plt.legend(loc="lower right", fontsize = "medium")
plt.xticks(rotation=0, horizontalalignment="center")
plt.yticks(rotation=0, horizontalalignment="right")
plt.show()
```
